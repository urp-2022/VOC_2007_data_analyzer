```python
def counting_train_data(category, class_type):
  f = open("/content/VOC/train/VOCdevkit/VOC2007/ImageSets/Main/"+category+"_"+class_type+".txt", 'r')
  cnt = 0

  while True:
    line = f.readline()
    if line:
      label = line.split(' ')[1]
      if not '-1' in label:
        cnt+=1
    else:
      break

  f.close()

  return cnt
```

```python
VOC_CLASSES = [
    'aeroplane', 'bicycle', 'bird', 'boat',
    'bottle', 'bus', 'car', 'cat', 'chair',
    'cow', 'diningtable', 'dog', 'horse',
    'motorbike', 'person', 'pottedplant',
    'sheep', 'sofa', 'train', 'tvmonitor'
]
```

```python
train_cnt = []

for i in range(len(VOC_CLASSES)):
  cnt = counting_train_data(VOC_CLASSES[i], 'train')
  train_cnt.append(cnt)
  print(VOC_CLASSES[i]+" "+str(cnt))

print(train_cnt)
```

```python
val_cnt = []

for i in range(len(VOC_CLASSES)):
  cnt = counting_train_data(VOC_CLASSES[i], 'val')
  val_cnt.append(cnt)
  print(VOC_CLASSES[i]+" "+str(cnt))

print(val_cnt)
```

```python
all_cnt = []

for i in range(len(VOC_CLASSES)):
  cnt = counting_train_data(VOC_CLASSES[i], 'trainval')
  all_cnt.append(cnt)
  print(VOC_CLASSES[i]+" "+str(cnt))

print(all_cnt)
```
